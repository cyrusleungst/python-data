# Categorical Plots

```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

### barplot()

```python
sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
```

### countplot()

```python
sns.countplot(x='sex', data=tips)
```

### box/whiskerplot()

```python
# Can use hue argument to add another categorical column
sns.boxplot(x='day',y='total_bill', data=tips, hue='smoker')
```

### violinplot()
This shows the distribution of quantitative data across several levels of categorical variables such that those distributions can be compared. It also features a kernel density estimation of the underlying distribution.

```python
sns.violinplot(x='day',y='total_bill', data=tips)
```

```python
sns.violinplot(x='day',y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill', data=tips, hue='sex', split=True)
```

### stripplot()
A stripplot will draw a scatterplot where one variable is categorical.

```python
sns.stripplot(x='day',y='total_bill', data=tips)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=False)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue='sex')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, hue='sex', dodge=True)
```

### swarmplot()
Swarmplot is similar to stripplot(), but the points are adjusted (only along the categorical axis) so that they don’t overlap. This also gives a better representation of the distribution of values.

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

### You can combine plots together

```python
sns.violinplot(x="tip", y="day", data=tips)
sns.swarmplot(x="tip", y="day", data=tips,color='black')
```

### catplot()
The most general form of a categorical plot. It can take in a kind parameter to adjust the plot type:

```python
sns.catplot(x="day", y="total_bill", data=tips, kind='bar')
```
