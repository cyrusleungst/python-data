# Regression

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### lmplot()
Displays linear models and to split up plots based on features

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'])
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o','v'], scatter_kws={'s':100})
```

### Assisting with grids
Allows for more variable seperation through rows and columns

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex')
```

### Aspect/Size ratios
Use height and aspect arguments to make figures easier to look at

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex', aspect=0.6, height=8)
```
