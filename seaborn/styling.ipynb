```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips)
```

```python
sns.countplot(x='sex',data=tips)
sns.despine(left=True)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='seismic')
```

### Size/Aspect
You can use matplotlib for this or just pass in height/aspect arguments

```python
plt.figure(figsize=(14,3))
sns.countplot(x='sex',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, height=3, aspect=4)
```

### set_context()
Allows for overriding of default parameters

```python
sns.set_context('poster')
sns.countplot(x='sex',data=tips)
```
