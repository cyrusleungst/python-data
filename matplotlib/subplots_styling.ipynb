# Subplotting and adding styling

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Subplots

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_xlabel('y')
axes.set_title('Title')
```

```python
# Add arguments in subplots() to change column and row size
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# 'axes' is an array of axes to plot on that can be iterated through
for ax in axes:
    ax.plot(x,y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Title')

fig
```

```python
# Use tight_layout() to prevent overlapping subplots/figures
# 'axes' is an array of axes to plot on that can be iterated through
fig, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x,y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Title')

fig
plt.tight_layout()
```

## Changing figure sizes and aspect ratios

```python
# figsize is the width and height of the figure in inches
# dpi is the dots per inch
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(15,3))

axes.plot(x,y, 'g')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title')
```

## Saving figures into different formats

```python
# Figures can saved in many formats: png. jpg, svg, pdf etc
fig.savefig('figure1.png', dpi=200)
```

## Legends and labels

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

# Use label keyword with legend() to add a legend to the figure
ax.plot(x, x**2, label='x**2')
ax.plot(x, x**3, label='x**3')
ax.legend()
```

## Line and marker style

```python
fig, ax = plt.subplots(figsize=(12,6))

# Use 'lw' to set line width and 'ls' to set line style
ax.plot(x, x+5, color="green", lw=3, linestyle='-')
ax.plot(x, x+6, color="green", lw=3, ls='-.')
ax.plot(x, x+7, color="green", lw=3, ls=':')

# marker size and colours
ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

## Setting plot range

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Use set_ylim and set x_lim to set range of axes. axis('tight') to get tight fitted axes ranges
axes[0].plot(x, x**2, x, x**3)
axes[0].set_title("default axes ranges")

axes[1].plot(x, x**2, x, x**3)
axes[1].axis('tight')
axes[1].set_title("tight axes")

axes[2].plot(x, x**2, x, x**3)
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title("custom axes range");
```
