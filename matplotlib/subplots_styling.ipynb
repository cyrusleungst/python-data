# Subplotting and adding styling

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Subplots

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_xlabel('y')
axes.set_title('Title')
```

```python
# Add arguments in subplots() to change column and row size
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# 'axes' is an array of axes to plot on that can be iterated through
for ax in axes:
    ax.plot(x,y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Title')

fig
```

```python
# Use tight_layout() to prevent overlapping subplots/figures
# 'axes' is an array of axes to plot on that can be iterated through
fig, axes = plt.subplots(nrows=1, ncols=2)
for ax in axes:
    ax.plot(x,y, 'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Title')

fig
plt.tight_layout()
```

## Changing figure sizes and aspect ratios

```python
# figsize is the width and height of the figure in inches
# dpi is the dots per inch
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(15,3))

axes.plot(x,y, 'g')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Title')
```
