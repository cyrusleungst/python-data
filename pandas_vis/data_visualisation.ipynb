## Plot Types

### Area

```python
df2.plot.area()
```

```python
# Use alpha argument to set transparency
df2.plot.area(alpha=0.5)
```

### Bar Plot

```python
df2.plot.bar()
```

```python
# stacked argument to get a stacked bar chart
df2.plot.bar(stacked=True)
```

### Scatter Plot

```python
df1.plot.scatter(x='A', y='B')
```

```python
# cmap argument to select colourmap and c to colour based off another coloumn value
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

```python
# s argument to change size based off another column value
df1.plot.scatter(x='A',y='B',s=df1['C']*50)
```

### Box Plot

```python
df2.plot.box()
```

### Hexagonal Bin Plot

```python
df = pd.DataFrame(np.random.randn(1000,2), columns=['a','b'])
```

```python
df.head()
```

```python
df.plot.hexbin(x='a', y='b', gridsize=25)
```
